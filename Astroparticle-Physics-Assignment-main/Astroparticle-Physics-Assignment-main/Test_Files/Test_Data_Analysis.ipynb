```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score

# Assume `data` is your feature matrix and `classes` is your target vector
# Replace `data` and `classes` with your actual data and class labels


# Convert X_train and X_test to NumPy arrays if they are lists
if isinstance(X_train, list):
    X_train = np.array(X_train)
if isinstance(X_test, list):
    X_test = np.array(X_test)
if isinstance(Y_train, list):
    Y_train = np.array(Y_train)
if isinstance(Y_test, list):
    Y_test = np.array(Y_test)

# Create an SVM classifier
svm_classifier = SVC(kernel='linear', random_state=0)

# Train the classifier on the training data
svm_classifier.fit(X_train, Y_train)

# Make predictions on the test data
Y_pred = svm_classifier.predict(X_test)

# Evaluate the classifier
accuracy = accuracy_score(Y_test, Y_pred)
report = classification_report(Y_test, Y_pred)

print(f"Accuracy: {accuracy}")
print("Classification Report:")
print(report)

# Define a function to plot decision boundaries
def plot_decision_boundaries(X, y, model, title='SVM Decision Boundaries'):
    h = .02  # step size in the mesh
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    plt.contourf(xx, yy, Z, alpha=0.8)
    plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k', marker='o')
    plt.xlabel('$dN_{e\mu}$')
    plt.ylabel('$\log_{10}(N_e)$')
    plt.title(title)
    plt.show()

# Plot decision boundaries for the training data
f2 = plt.figure()
plot_decision_boundaries(X_train, Y_train, svm_classifier, title='SVM Decision Boundaries (Training Data)')
f2.savefig('SVM-linear-training.pdf',bbox_inches='tight')
# Plot decision boundaries for the test data
f3 = plt.figure()
plot_decision_boundaries(X_test, Y_test, svm_classifier, title='SVM Decision Boundaries (Test Data)')
f3.savefig('SVM-linear-test.pdf',bbox_inches='tight')
```
